```python
import pandas as pd
import matplotlib as plt
col_names = ["qaccver","saccver","pident","length","mismatch","gapopen","qstart","qend","sstart","send","evalue","bitscore"]
# read the CSV file into a pandas DataFrame
blast_results = pd.read_csv('sample0_test_outputs.txt', sep=',', header=0,names=col_names)

# count the number of hits per species
query_counts = blast_results['qaccver'].value_counts()
```

```python
display(blast_results)
display(query_counts)
```

```python
# # count the occurrences of each item
# counts = blast_results['qaccver'].value_counts()

# # select the items that occur 2 times or more
# threshold = 10000
# items = counts[counts >= threshold].index.tolist()

# # filter the DataFrame to only include the selected items
# filtered_df = blast_results[blast_results['qaccver'].isin(items)]
# display(filtered_df)

# group the DataFrame by the sample column and calculate the mean of each group
grouped = blast_results.groupby('qaccver').mean()
display(grouped)
eval_samples = grouped[['evalue']]
display(eval_samples)
```

```python
figsize_setting = (10,6)
# plot the results as a bar chart
ax = eval_samples.plot(kind='bar', figsize=figsize_setting, fontsize=12)
ax.set_xlabel('Transcript', fontsize=12)
ax.set_ylabel('Average E-Value', fontsize=12)
ax.set_title('Average E-Value vs Query Transcript', fontsize=14)
ax.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
```
